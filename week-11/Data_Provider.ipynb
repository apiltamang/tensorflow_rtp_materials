```python
import numpy as np
```

```python
images_list = []
labels_list = []
for filename in glob.glob('Product_Images/grayscale/*.jpg'): #assuming gif
    image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    images_list.append(image)
    #lastSlash = filename.rfind('/')
    #dest = filename[lastSlash+1:]
    #cv2.imwrite('Product_Images/grayscale/'+dest, image)
print(len(images_list))
```
